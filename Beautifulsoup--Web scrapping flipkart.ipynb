# 2. Multiple products scrapping(soup.findALL( 'div',attrs={'class': '         '}))

```python
#Only Connection done by selenium,will work with beautifulsoup here.
from selenium import webdriver
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
driver=webdriver.Chrome('C:\\chromedriver.exe')
```

```python
#Creating two enpty list where we will store our data.
products=[]
prices=[]
#Specifying link of the website
driver.get('https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=a626f5a8-4310-498d-a54f-f3b74433e3bb&as-searchtext=lap')
```

# 1.Extracting names(which will be appended to blank products[ ] list made above)

```python
#extracting html files of the given link above and feeding it into beautifulsoup for further works.
content=driver.page_source 
soup=BeautifulSoup(content)
names=soup.findAll('div',attrs={'class':'_3wU53n'})
print(len(names))
```

```python
names
```

```python
for name in names:
    products.append(name.text)
```

# 1.Extracting price(which will be appended to blank prices[ ] list made above.)

```python
amount=soup.find_all('div',attrs={'class':'_1vC4OE _2rQ-NK'})
print(len(amount))
```

```python
amount[0]
```

```python
for price in amount:
    prices.append(price.text)
```

```python
prices
```

```python
df=pd.DataFrame({'ProductName':products,'price':prices})
```

```python
df
```

# Store the data in required format
